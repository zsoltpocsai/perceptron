# Learning an AND gate

```python
import sys
import matplotlib.pyplot as plt
import numpy as np
```

```python
sys.path.append('../')
```

```python
from perceptron import Perceptron
```

## Data

```python
X = [[0, 0],
     [1, 0],
     [0, 1],
     [1, 1]]
Y = [0, 0, 0, 1]
data = { 'x': X, 'y': Y }
```

```python
p = Perceptron(2)
```

## Before training

```python
w, b = p.get_params()
print(f"Weights: {w}, Bias: {b}")
```

```python
print(f"Error: {p.eval(data)}")
```

```python
# plotting

x_0 = [X[i][0] for i in range(len(X)) if Y[i] == 0]
y_0 = [X[i][1] for i in range(len(X)) if Y[i] == 0]
x_1 = [X[i][0] for i in range(len(X)) if Y[i] == 1]
y_1 = [X[i][1] for i in range(len(X)) if Y[i] == 1]

p_x, p_y = p.line(np.arange(0.1, 1, 0.1))

plt.plot(x_0, y_0, 'xb')
plt.plot(x_1, y_1, 'og', fillstyle='none')
plt.plot(p_x, p_y)
```

```python
for i in range(10):
    p.train(data)
```

## After training

```python
w, b = p.get_params()
print(f"Weights: {w}, Bias: {b}")
```

```python
print(f"Error: {p.eval(data)}")
```

```python
# plot after training

p_x, p_y = p.line(np.arange(0.1, 1, 0.1))

plt.plot(x_0, y_0, 'xb')
plt.plot(x_1, y_1, 'og', fillstyle='none')
plt.plot(p_x, p_y)
```
